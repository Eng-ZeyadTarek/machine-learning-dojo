### import required packages

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import GaussianNB
from mlxtend.plotting import plot_decision_regions
```

### read data

```python
#read *binclass.txt* file
data= pd.read_csv('binclass.txt', sep=",", header=None, names=["a", "b", "c"])
```

### divide them to 2 input features and label column

```python
#X is two columns and y is one column
X = data.iloc[:,0:2].values
Y = data.iloc[:,2].values
```

### plot input features with label 1 in blue and input features with label -1 *(optional)*

```python
from collections import Counter
#print(counter)
for label, _ in counter.items():
    row_ix = np.where(Y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

### bulid and train **Gaussian Naive Bayes** model

```python
model_nb = GaussianNB()
model_nb.fit(X,Y)
```

### draw decision boundary

```python
plot_decision_regions(X, Y, clf = model_nb, legend = 2)
#title
plt.title("Decision boundary using gaussian naive Bayes (Train)")
#X label
plt.xlabel("mean_perimeter")
```
