```python
#Import the required modules, numpy for calculation, and Matplotlib for drawing
import numpy as np
import matplotlib.pyplot as plt
#This code is for jupyter Notebook only
%matplotlib inline

# define data, and change list to array
x = [3,21,22,34,54,34,55,60,67,89,99, 120, 130, 145]
x = np.array(x)
y = [1,10,14,34,44,36,22,50,67,79,90, 101, 115, 121]
y = np.array(y)

#Show the effect of a scatter plot
plt.scatter(x,y)
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.title("Data scatter plot")
```

```python
#The basic linear regression model is wx+ b, and since this is a two-dimensional space, the model is ax+ b
def linear_reg_model(a, b, x):

  return a*x + b



#The most commonly used loss function of linear regression model is the loss function of mean variance difference
def MSE(a, b, x, y):
    num = len(x)
    prediction=linear_reg_model(a,b,x)
    return (0.5/num) * (np.square(prediction-y)).sum()


#The optimization function mainly USES partial derivatives to update two parameters a and b gradient decent
def gradient_decent(a,b,x,y,Lr):
    num = len(x)
    prediction = linear_reg_model(a,b,x)
    #Update the values of A and B by finding the partial derivatives of the loss function on a and b
    da = (1.0/num) * ((prediction -y)*x).sum()
    db = (1.0/num) * ((prediction -y).sum())
    a = a - Lr*da
    b = b - Lr*db
    return a, b
```

```python
#iterated function, return a and b
error=[]
def iterate(a,b,x,y,times):
    for i in range(times):
        a,b = gradient_decent(a,b,x,y, 0.00001)
        loss = MSE(a, b, x, y)
        error.append(loss)
    return a,b,error
```

```python
#Initialize parameters and display
a = np.random.rand(1)
print(f"initial value of w1 is {a[0]:.4f}")
b = np.random.rand(1)
print(f"initial value of w2 is {a[0]:.4f}")
#For the first iteration, the parameter values, losses, and visualization after the iteration are displayed
a,b, error= iterate(a,b,x,y,500)
prediction=linear_reg_model(a,b,x)
loss = MSE(a, b, x, y)
print(f"w1 is {a[0]:.3f} and w2 is {b[0]:.3f} and the loss is {loss:.3f}.")
plt.scatter(x,y)
plt.plot(x,prediction)
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.title("linear regression model")
```
